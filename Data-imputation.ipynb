```python
index_nan = df.index[(df.Age.isna()==True) & (df.Cabin.isna()==True)]
```

## KNN Imputation

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=2)
KNN_df = imputer.fit_transform(df.drop(['Name','Sex', 'Ticket', 'Cabin', 'Embarked'],axis=1))
```

```python
pd.DataFrame(df,columns = df.drop(['Name','Sex', 'Ticket', 'Cabin', 'Embarked'],axis=1).columns).iloc[index_nan]
```

### OUTPUT

```python
pd.DataFrame(KNN_df,columns = df.drop(['Name','Sex', 'Ticket', 'Cabin', 'Embarked'],axis=1).columns).iloc[index_nan]
```

# MICE

```python
from impyute.imputation.cs import mice
features = df.drop(['Name','Sex', 'Ticket', 'Cabin', 'Embarked'], axis=1)
trans_mise = mice(features.values)
pd.DataFrame(df, columns=df.drop(['Name','Sex', 'Ticket', 'Cabin', 'Embarked'],axis=1).columns).iloc[index_nan]
```

### OUTPUT

```python
pd.DataFrame(trans_mise, columns=df.drop(['Name','Sex', 'Ticket', 'Cabin', 'Embarked'],axis=1).columns).iloc[index_nan]
```
